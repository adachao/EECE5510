```
# Convolve the image `Marquette.jpg' with a 15x15 mean filter using the following padding strategies: Zero padding, Nearest neighbor, reflect. Visually compare the results.
import cv2
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt
#read image and convert it into an ndarray
img = cv2.imread('Marquette.jpg')
#convert img to grayscale
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#set 15x15 mean filter
k = np.ones((15,15))/225
#Zero padding, Nearest neighbor, reflect
output_image1 = scipy.ndimage.filters.convolve(img,k,mode = 'constant', cval = 0.0)
output_image2 = scipy.ndimage.filters.convolve(img,k,mode = 'nearest')
output_image3 = scipy.ndimage.filters.convolve(img,k,mode = 'reflect')
#plot images
plt.imshow(output_image1,'gray')
plt.show()
#plot images
plt.imshow(output_image2,'gray')
plt.show()
#plot images
plt.imshow(output_image3,'gray')
plt.show()
```

```
# Create a numpy array with the same size as the image `Marquette.jpg' containing Gaussian noise with mean 30 and variance 6 and add the two images (make sure the pixels in the resulting image are in the range [0, 255])
import cv2
import numpy as np
import matplotlib.pyplot as plt
#read image and convert it into an ndarray
img = cv2.imread('Marquette.jpg')
#convert img to grayscale
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#plot image
plt.imshow(img,'gray')
plt.show()

#set mean 30 and variance 6
mean = 30
variance = 6
k = 5
#apply Gaussian noise with mean 30 and variance 6
Gaussian_noise = k*np.random.normal(mean, np.sqrt(variance), size = img.shape)
#add the two images
output_image4 = img+Gaussian_noise
#plot image
plt.imshow(output_image4,'gray')
plt.show()
```

```
# Apply a Gaussian filter with mean 0 and variance 6 to the image and display the resulting image.
import cv2
#import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt
#set mean 0 and variance 6
variance = 6
#Apply a Gaussian filter
output_image5 = scipy.ndimage.filters.gaussian_filter(output_image4, sigma = np.sqrt(variance), mode='reflect')
#plot image
plt.imshow(output_image5,'gray')
plt.show()
```

```
# Apply a mean filter to the image corrupted with noise. Experiment with the filter size to optimize the results.
import cv2
#import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt
#Experiment with the different filter size
for n in range(1,10):
  k = np.ones((n,n))/(n*n)
# Apply a mean filter to the image corrupted with noise
  output_image6 = scipy.ndimage.filters.convolve(output_image4, k)
#plot image
  plt.imshow(output_image6,'gray')
  plt.show()
```

```
# Create a numpy array with the same size as the image `Marquette.jpg' containing impulse noise where the probability that a pixel is corrupted is 10% and add the two images (make sure the pixels in the resulting image are in the range [0, 255])
import cv2
import numpy as np
import matplotlib.pyplot as plt
#read image and convert it into an ndarray
img = cv2.imread('Marquette.jpg')
#convert img to grayscale
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#plot image
plt.imshow(img,'gray')
plt.show()

#pixel is corrupted is 10%
pro = 0.1
k = 100
#add impulse noise
impulse_noise = pro*k*np.random.randn(img.shape[0], img.shape[1])/(1+pro)
#add the two images
output_image7 = img+impulse_noise
#plot
plt.imshow(output_image7,'gray')
plt.show()
```

```
# Apply a mean filter to the image corrupted with noise. Experiment with the filter size to optimize the results.
import cv2
#import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt
#Experiment with the different filter size
for n in range(1,10):
  k = np.ones((n,n))/(n*n)
#Apply a mean filter to the image corrupted with noise
  output_image8 = scipy.ndimage.filters.convolve(output_image7, k)
  plt.imshow(output_image8,'gray')
  plt.show()
```

```
# Apply the Sobel edge detector and the Canny edge detector to the image `Marquette.jpg' and visually compare the results.
import cv2
from scipy import ndimage
from matplotlib import pyplot as plt
#from scipy.ndimage import sobel

#read image and convert it into an ndarray
img = cv2.imread('Marquette.jpg')
#convert img to grayscale
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#plot
plt.imshow(img,'gray')
plt.show()

#Sobel edge detector
img_sobel = ndimage.sobel(img)\
#plot
plt.imshow(img_sobel,'gray')
plt.show()

#Canny edge detector
img_canny = cv2.Canny(img, 100, 200)
#plot
plt.imshow(img_canny,'gray')
plt.show()
```
