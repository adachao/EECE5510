```
# Convolve the image `Marquette.jpg' with a 15x15 mean filter using the following padding strategies: Zero padding, Nearest neighbor, reflect. Visually compare the results.
import cv2
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage

!wget "https://github.com/Henry-Medeiros/images/raw/master/Marquette.jpg"
# Opening the image using cv2.
img = cv2.imread('Marquette.jpg')
# Converting the image to grayscale.
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


# Initializing the filter of size 15 by 15.
# The filter is divided by 225 for normalization.
k = np.ones((15,15))/225
# performing convolution
b = scipy.ndimage.filters.convolve(img, k)
#Add padding strategies to output image
img_zero_pad = scipy.ndimage.filters.convolve(img,k,mode = 'constant',cval=0.0)
img_reflect_pad = scipy.ndimage.filters.convolve(img,k,mode = 'reflect')
img_nearest_pad = scipy.ndimage.filters.convolve(img,k,mode = 'nearest')
#plot the images
plt.imshow(img_zero_pad, 'gray')
plt.show()
plt.imshow(img_reflect_pad,'gray')
plt.show()
plt.imshow(img_nearest_pad,'gray')
plt.show()
```

```
# Create a numpy array with the same size as the image `Marquette.jpg' containing Gaussian noise with mean 30 and variance 6 and add the two images (make sure the pixels in the resulting image are in the range [0, 255])
import cv2
import matplotlib.pyplot as plt
import numpy as np

!wget "https://github.com/Henry-Medeiros/images/raw/master/Marquette.jpg"
# Opening the image using cv2.
img = cv2.imread('Marquette.jpg')
# Converting the image to grayscale.
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#plot the image
plt.imshow(img, 'gray')
plt.show()

# set mean and variance
mean = 30
variance = 6
#contain gaussian noise 
gauss = np.random.normal(mean,np.std(variance), size=img.shape)
img_gauss = (255*(img+gauss)).astype(np.uint8)
#plot the image
plt.imshow(img_gauss, 'gray')
plt.show()
```

```
# Apply a Gaussian filter with mean 0 and variance 6 to the image and display the resulting image.
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter

!wget "https://github.com/Henry-Medeiros/images/raw/master/Marquette.jpg"
# Opening the image using cv2.
img = cv2.imread('Marquette.jpg')
# Converting the image to grayscale.
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# set mean and variance
mean = 0
variance = 6
#apply gaussian filter
img_gauss = gaussian_filter(img, sigma = np.std(variance),mode = 'reflect')
#plot the image
plt.imshow(img_gauss, 'gray')
plt.show()
```

```
# Apply a mean filter to the image corrupted with noise. Experiment with the filter size to optimize the results.
import cv2
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage 

!wget "https://github.com/Henry-Medeiros/images/raw/master/Marquette.jpg"
# Opening the image using cv2.
img = cv2.imread('Marquette.jpg')
# Converting the image to grayscale.
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Initializing the filter of size n by n.
# The filter is divided by n*n for normalization.
#I will set a small range because of the efficiency
for n in range(1,8):
  k = np.ones((n,n))/(n*n)
#perform convolution
  img_mean = scipy.ndimage.filters.convolve(img, k)
#plot the image
  print(n)
  plt.imshow(img_mean, 'gray')
  plt.show()
```

```
#Create a numpy array with the same size as the image `Marquette.jpg' containing impulse noise where the probability that a pixel is corrupted is 10% and 
#add the two images (make sure the pixels in the resulting image are in the range [0, 255])
import cv2
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage

!wget "https://github.com/Henry-Medeiros/images/raw/master/Marquette.jpg"
# Opening the image using cv2.
img = cv2.imread('Marquette.jpg')
# Converting the image to grayscale.
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#plot the image
plt.imshow(img, 'gray')
plt.show()

#a pixel is corrupted is 10%
p = 0.1

#containing impulse noise
noise = np.random.normal(img.shape[0], img.shape[1])/(1+p)
noise_img = img+noise
noise_img=(255*noise_img).astype(np.uint8)
#plot the image
plt.imshow(noise_img,'gray')
plt.show()
```

```
# Apply a mean filter to the image corrupted with noise. Experiment with the filter size to optimize the results.
import cv2
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage 

# Initializing the filter of size n by n.
# The filter is divided by n*n for normalization.
#I will set a small range because of the efficiency
for n in range(1,8):
  k = np.ones((n,n))/(n*n)
#perform convolution
  img_mean = scipy.ndimage.filters.convolve(noise_img, k)
#plot the image
  print(n)
  plt.imshow(img_mean, 'gray')
  plt.show()
```

```
# Apply the Sobel edge detector and the Canny edge detector to the image `Marquette.jpg' and visually compare the results.
import cv2
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage

# Opening the image using cv2.
img = cv2.imread('Marquette.jpg')
# Converting the image to grayscale.
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#plot the image
plt.imshow(img, 'gray')
plt.show()

#perform sobel filter
img_sobel = scipy.ndimage.sobel(img)
#plot the image
plt.imshow(img_sobel, 'gray')
plt.show()

#perform canny filter
img_canny = cv2.Canny(img, 100, 200)
#plot the image
plt.imshow(img_canny, 'gray')
plt.show()
```
